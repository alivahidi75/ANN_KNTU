**Plot Noise Levels**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


mean = 0.0
std = 0.1  
N = 10000 

noise1 = np.random.normal(mean, 0.008, N)
noise2 = np.random.normal(mean, 0.01, N)
noise3 = np.random.normal(mean, 0.02, N)
noise4 = np.random.normal(mean, 0.08, N)
noise5 = np.random.normal(mean, 0.1, N)
noise6 = np.random.normal(mean, 0.2, N)

sns.kdeplot(noise1, shade=False, color='blue', label="std = 0.008")
sns.kdeplot(noise2, shade=False, color='red', label="std = 0.01")
sns.kdeplot(noise3, shade=False, color='green', label="std = 0.02")
sns.kdeplot(noise4, shade=False, color='purple', label="std = 0.08")
sns.kdeplot(noise5, shade=False, color='black', label="std = 0.1")
sns.kdeplot(noise6, shade=False, color='orange', label="std = 0.2")

plt.title(f"Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

**Bar Plots**

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['Clean Data', 'std = 0.008', 'std = 0.01', 'std = 0.02','std = 0.08', 'std = 0.1', 'std = 0.2']


group1 = [99.603, 99.473, 99.435, 99.397, 98.319, 97.312, 92.652]
group2 = [98.823, 98.516, 98.671, 98.483, 97.162, 95.593, 92.068]
group3 = [98.446, 98.098, 98.25, 97.973, 96.525, 95.636, 92.299]
group4 = [99.383, 99.347, 99.209, 99.132, 97.28, 96.816, 91.851]


x = np.arange(len(labels)) 
width = 0.3           

plt.figure(figsize=(10, 5))

plt.bar(x - width/2, group1, width, label='Train-Rough', color='skyblue')


plt.bar(x - width/4, group2, width, label='Train-RBF', color='orange')

plt.bar(x + width/4, group3, width, label='Train-Rough-RBF', color='blue')


plt.bar(x + width/2, group4, width, label='Train-MLP', color='red')

plt.xticks(x, labels)


plt.xlabel('Scenarios')
plt.ylabel('Accuracy')
plt.title('Accuracy of clean and noisy data')
plt.legend()

plt.show()

```
